```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
aceleracion_mpu_x = pd.read_csv('mpu/mpu_ascensor_x.csv')
aceleracion_celular_x = pd.read_csv('celular/celular_ascensor_x.csv')

aceleracion_mpu_y = pd.read_csv('mpu/mpu_ascensor_y.csv')
aceleracion_celular_y = pd.read_csv('celular/celular_ascensor_y.csv')

aceleracion_mpu_z = pd.read_csv('mpu/mpu_ascensor_z.csv')
aceleracion_celular_z = pd.read_csv('celular/celular_ascensor_z.csv')
```

```python
x = aceleracion_mpu_x.Medicion
mpu_x = aceleracion_mpu_x.Aceleracion_X
celular_x = aceleracion_celular_x.gFx

plt.plot(x, mpu_x, label='MPU')
plt.plot(x, celular_x, label='Celular')
plt.ylim([0.8, 1.2])
plt.legend()

plt.show()
```

```python
x = aceleracion_mpu_y.Medicion
mpu_y = aceleracion_mpu_y.Aceleracion_Y
celular_y = aceleracion_celular_y.gFy

plt.plot(x, mpu_y, label='MPU')
plt.plot(x, celular_y, label='Celular')
plt.ylim([0.5, 1.5])
plt.legend()

plt.show()
```
