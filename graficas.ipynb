```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bmp_naranja = pd.read_csv('datos_bmp_naranja.csv')
bmp_sin_naranja = pd.read_csv('datos_bmp_sin_naranja.csv')
camara_presion = pd.read_csv('camara_presion.csv')
mpu_reposo = pd.read_csv('mpu_reposo.csv')

#archivo = pd.read_csv('datos.csv')
x = bmp_naranja.Medicion
```

```python
mpu_reposo_x = mpu_reposo.Medicion
mpu_reposo_y_x = mpu_reposo.Aceleracion_X
mpu_reposo_y_y = mpu_reposo.Aceleracion_Y
mpu_reposo_y_z = mpu_reposo.Aceleracion_Z

plt.plot(mpu_reposo_x, mpu_reposo_y_x, label='X')
plt.plot(mpu_reposo_x, mpu_reposo_y_y, label='Y')
plt.plot(mpu_reposo_x, mpu_reposo_y_z, label='Z')
plt.legend()

plt.show()
```

```python
y_naranja_redondeada = bmp_naranja.Altitud_redondeada
y_naranja_sin_redondear = bmp_naranja.Altitud

plt.title('BMP Naranja')

plt.plot(x, y_naranja_redondeada, 'o-', label='redondeado', color='green')
plt.plot(x, y_naranja_sin_redondear, 'o-', label='sin redondear', color='blue')
plt.legend()

plt.xlabel('Mediciones')
plt.ylabel('BMP Naranja')
plt.show()
```

```python
y_sin_naranja_redondeada = bmp_sin_naranja.Altitud_redondeada
y_sin_naranja_sin_redondear = bmp_sin_naranja.Altitud

plt.title('BMP sin naranja')

plt.plot(x, y_sin_naranja_redondeada, 'o-', label='redondeado')
plt.plot(x, y_sin_naranja_sin_redondear, 'o-', label='sin redondear')
plt.legend()

plt.xlabel('Mediciones')
plt.ylabel('Altitud')
plt.show()
```
