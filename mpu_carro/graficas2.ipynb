```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
mpu = pd.read_csv('mpu_2.csv')
celular = pd.read_csv('celular2.csv')
```

```python
x = mpu.Medicion
mpu_x = mpu.Aceleracion_X
mpu_y = mpu.Aceleracion_Y
mpu_z = mpu.Aceleracion_Z

plt.title('mpu')
# plt.plot(x, mpu_x, label='X')
plt.plot(x, mpu_y, label='Y')
# plt.plot(x, mpu_z, label='Z')
plt.xlim([400, 430])

plt.legend()
plt.show()
```
