```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mpu = pd.read_csv('mpu.csv')
celular = pd.read_csv('celular.csv')
```

```python
x = mpu.Medicion
mpu_x = mpu.Aceleracion_X
mpu_y = mpu.Aceleracion_Y
mpu_z = mpu.Aceleracion_Z

plt.title('mpu')
plt.plot(x, mpu_x, label='X')
plt.plot(x, mpu_y, label='Y')
plt.plot(x, mpu_z, label='Z')
plt.legend()
plt.show()
```

```python
x = celular.time
celular_x = celular.gFx
celular_y = celular.gFy
celular_z = celular.gFz

plt.title('celular')
plt.plot(x, celular_x, label='X')
plt.plot(x, celular_y, label='Y')
plt.plot(x, celular_z, label='Z')
plt.legend()
plt.show()
```
