```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
aceleracion_mpu = pd.read_csv('mpu_reposo.csv')
aceleracion_celular = pd.read_csv('nuevo_mpu_reposo_celular.csv')
```

```python
x = aceleracion_mpu.Medicion

mpu_x = aceleracion_mpu.Aceleracion_X
mpu_y = aceleracion_mpu.Aceleracion_Y
mpu_z = aceleracion_mpu.Aceleracion_Z

celular_x = aceleracion_celular.gFx
celular_y = aceleracion_celular.gFy
celular_z = aceleracion_celular.gFz
```

```python
plt.plot(x, mpu_x, label='X MPU')
plt.plot(x, mpu_y, label='Y MPU')
plt.plot(x, mpu_z, label='Z MPU')
plt.plot(x, celular_x, label='X Celular')
plt.plot(x, celular_y, label='Y Celular')
plt.plot(x, celular_z, label='Z Celular')
plt.title('MPU y Celular')
plt.legend()
plt.show()
```

```python
plt.plot(x, mpu_x, label='X MPU')
plt.plot(x, celular_x, label='X Celular')
plt.title('MPU y Celular X')

plt.ylim([0, 1])

plt.legend()
plt.show()
```

```python
plt.plot(x, mpu_y, label='Y MPU')
plt.plot(x, celular_y, label='Y Celular')
plt.title('MPU y Celular Y')
plt.ylim([0, 1])
plt.legend()
plt.show()
```

```python
plt.plot(x, mpu_z, label='Z MPU')
plt.plot(x, celular_z, label='Z Celular')
plt.title('MPU y Celular Z')
plt.ylim([0, 1.2])
plt.legend()
plt.show()
```
